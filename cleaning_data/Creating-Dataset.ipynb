```python
import pandas as pd
import numpy as np
import random
```

```python
#Limpieza del DATASET
df = pd.read_csv('realtor-data.csv')
df = df.dropna()
df = df.reset_index(drop=True)
df.to_csv('realtor-data-clean.csv', index=False)
df
```

```python
# Función para verificar tipos de datos
def verificar_tipos_de_datos(dataframe):
    for columna in dataframe.columns:
        tipos_de_datos = dataframe[columna].apply(type)
        tipos_unicos = tipos_de_datos.unique()
        if len(tipos_unicos) == 1:
            print(f'La columna {columna} contiene solo valores del tipo {tipos_unicos[0]}.')
        else:
            print(f'La columna {columna} contiene múltiples tipos de datos: {tipos_unicos}.')

verificar_tipos_de_datos(df)
```

```python
#Crea un DataFrame que contenga todas las ciudades sin repetición
ciudades_estados_unicos = df[['city', 'state']].drop_duplicates()
df_cities = pd.DataFrame({"City": ciudades_estados_unicos['city'], "State": ciudades_estados_unicos['state']})

#Crea un DataFrame que contenga todos los estados sin repetición
states = df['state'].unique()
df_states = pd.DataFrame({"State": states})
```

```python
df['bed'] = df['bed'].astype(int)
df['bath'] = df['bath'].astype(int)
df['zip_code'] = df['zip_code'].astype(int)
df
```

```python
#Drop duplicates in df
df = df.drop_duplicates()
df
```

```python
#Search state and year in df
def search_state_year(state, year):
    df_state_year = df.loc[(df['state'] == state) & (df['year'] == year)]
    return df_state_year
```

```python
# Define la cantidad de personas que deseas en tu dataset
cantidad_personas = 50000

# Crea un DataFrame con datos aleatorios de personas
df_personas = pd.DataFrame({
    'ID': range(1, cantidad_personas + 1),
    'Name': ['Person' + str(i) for i in range(1, cantidad_personas + 1)],
    'Age': [random.randint(18, 65) for _ in range(cantidad_personas)],
    'Gender': [random.choice(['Male', 'Female']) for _ in range(cantidad_personas)],
})
```

```python
df_personas
```

```python
df_houses = df
personas = df_personas

#Añadir columna IDpersona a df2
IDs = df_personas['ID'].tolist()
random.shuffle(IDs)

#Asignar valores de IDs a df2 en la columna IDpersona de forma aleatoria
df_houses['ownerID'] = [random.choice(df_personas['ID']) for i in range(len(df_houses))]
df_houses
```
