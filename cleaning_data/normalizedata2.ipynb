```python
import pandas as pd
import numpy as np
import json
import csv
import re
import math
```

```python
# Función para eliminar comas dentro de comillas dobles
def remove_commas_inside_quotes(match):
    return match.group(0).replace(',', '')

# Abre el archivo CSV de entrada para lectura
with open('data2.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)

    # Abre un nuevo archivo CSV de salida para escritura
    with open('temp2.csv', 'w', newline='') as outfile:
        writer = csv.writer(outfile)

        # Itera a través de las filas del archivo CSV de entrada
        for row in reader:
            cleaned_row = []

            for col in row:
                # Utiliza una expresión regular para encontrar y reemplazar comas dentro de comillas dobles
                col = re.sub(r'[^"]*', remove_commas_inside_quotes, col)
                cleaned_row.append(col)

            # Escribe la fila limpiada en el nuevo archivo CSV
            writer.writerow(cleaned_row)
```

```python
df = pd.read_csv("temp2.csv", quoting=3, delimiter=',')
df
```

```python
#Search for El Dorado County
df = df[df['RegionName'] == 'El Dorado County']
df
```

```python
# Load JSON data from a file
with open('fips_info.json', 'r') as json_file:
    data = json_file.read()

    # Parse JSON into a Python dictionary
    json_data = json.loads(data)

    # Create a DataFrame from the dictionary
    df_fips = pd.DataFrame(json_data)

df_fips
```

```python
## Merge dataframes by county name

new_df = pd.merge(df, df_fips,  how='left', left_on=['RegionName','StateName'], right_on = ['area_name','state'])
```

```python
new_df
```

```python
## Drop unnecessary columns
new_df = new_df.drop(['area_name', 'Metro', 'StateCodeFIPS', 'MunicipalCodeFIPS', 'bachelorsOrHigher','state','RegionType'], axis=1)
new_df
```

```python
## Move fips column to the beginning
first_column = new_df.pop('fips')
new_df.insert(3, 'FipsCode', first_column)
new_df
```
