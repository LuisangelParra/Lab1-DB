```python
import csv
import re

# Función para eliminar comas dentro de comillas dobles
def remove_commas_inside_quotes(match):
    return match.group(0).replace(',', '')

# Abre el archivo CSV de entrada para lectura
with open('data.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)

    # Abre un nuevo archivo CSV de salida para escritura
    with open('temp.csv', 'w', newline='') as outfile:
        writer = csv.writer(outfile)

        # Itera a través de las filas del archivo CSV de entrada
        for row in reader:
            cleaned_row = []

            for col in row:
                # Utiliza una expresión regular para encontrar y reemplazar comas dentro de comillas dobles
                col = re.sub(r'"[^"]*"', remove_commas_inside_quotes, col)
                cleaned_row.append(col)

            # Escribe la fila limpiada en el nuevo archivo CSV
            writer.writerow(cleaned_row)
```

```python
df = pd.read_csv("temp.csv", quoting=3, delimiter=',')
df['RegionID'] = df['RegionID'].str.replace('"', '')
df['2023-09-30'] = df['2023-09-30'].str.replace('"', '')
df
```
